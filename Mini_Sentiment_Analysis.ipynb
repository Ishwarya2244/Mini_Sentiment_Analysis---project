## **Importing required libraries**

```python
import pandas as pd
import numpy as np
#“From the sklearn library
#go to the model section
# and take only train_test_split
from sklearn.model_selection import train_test_split
#Converts text → numbers
from sklearn.feature_extraction.text import TfidfVectorizer
#“From the sklearn library
# go to the linear_model section
# and take only LogisticRegression”
from sklearn.linear_model import LogisticRegression
#Checks how good your model is
#accuracy_score → % correct predictions
#classification_report → precision, recall, F1-score
from sklearn.metrics import accuracy_score,classification_report
```

# **Created a data**

```python
data={"review":["This movie is amazing","I loved the acting","What a fantastic film","I hated this movie","This film is terrible","Worst movie ever"],"sentiment":[1,1,1,0,0,0]}
df=pd.DataFrame(data)
print(df)
```

# **split input and output**

```python
#If I ask you:

#“This movie is amazing”

#And I want the computer to answer:

#Positive or Negative?
#X = Question (Input),What you GIVE to the model, Here: the review text
# y = Answer (Output),What you WANT the model to predict, Here: the sentiment
X=df['review']
Y=df['sentiment']
```

# **splitting training and testing**

```python
#You can use:test_size = 0.2 → 80% train, 20% test (most common),test_size = 0.3 → 70% train, 30% test
#test_size = 0.25 → 75% train, 25% test
#Every time you run code, Python randomly shuffles data.
#You can uuse:0,1,..100

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=30)
```

# **convert text to numbers**

```python
#TfidfVectorizer(): convert text to numbers
#fit_transform(x_train): fit: learn vocabulary from training data, transform:convert text to numbers
#never fit on test data bcz it uses the same learned vocabulary.

vectorizer=TfidfVectorizer()
X_train_tfidf=vectorizer.fit_transform(X_train)
X_test_tfidf=vectorizer.transform(X_test)
```

# **Train the model**

```python
#LogisticRegression() → creates a machine learning model
#fit() → tells the model:
#Look at the training data (X_train_tfidf)
#Compare it to the correct answers (y_train)
#Learn patterns so it can predict new data


model =LogisticRegression()
model.fit(X_train_tfidf,Y_train)
```

# **make predictions**

```python
#make predictions on test data
#X_test_tfidf → test reviews converted to numbers
#model.predict() → uses the patterns learned in Step 7
#Output (y_pred) → predicted sentiment (0 = Negative, 1 = Positive)

Y_pred=model.predict(X_test_tfidf)
```

```python
#
print("Accuracy:",accuracy_score(Y_test,Y_pred))
print("\nClassification Report:\n",classification_report(Y_test,Y_pred))
```
